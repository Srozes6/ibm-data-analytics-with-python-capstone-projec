### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
# Get the top five databases respondents want to work with
top_databases = df['DatabaseWantToWorkWith'].value_counts().head(5)

# Create the pie chart
plt.figure()
plt.pie(top_databases, labels=top_databases.index, autopct='%1.1f%%', startangle=140)
plt.title('Top Five Databases Respondents Want to Work With')
plt.axis('equal')

plt.show()

# The pie chart indicates that interest is concentrated in a small number of databases, with the top five accounting for most of the responses. This suggests respondents strongly favor widely adopted and industry-relevant databases over less common options. Overall, the results highlight a clear preference for a core set of database technologies seen as valuable for career growth.
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
# Get the top five developer roles
top_dev_types = df['DevType'].value_counts().head(5)

# Create the pie chart
plt.figure()
plt.pie(top_dev_types, labels=top_dev_types.index, autopct='%1.1f%%', startangle=140)
plt.title('Top Five Developer Roles Among Respondents')
plt.axis('equal')

plt.show()

# The pie chart illustrates the distribution of the top five developer roles among survey respondents. A small number of roles make up the majority of responses, indicating that certain developer positions are more common within the survey population. This suggests that the dataset is dominated by a few core roles, while other developer types are less represented.
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
##Write your code here

top_os = df['OpSysPersonal use'].value_counts().head(5)

# Create the pie chart
plt.figure()
plt.pie(top_os, labels=top_os.index, autopct='%1.1f%%', startangle=140)
plt.title('Top Operating Systems Used by Respondents (Professional Use)')
plt.axis('equal')

plt.show()

#  pie chart shows that professional operating system usage is dominated by a few platforms, with one or two operating systems accounting for the majority of responses. This indicates a strong preference for widely supported and industry-standard operating systems among developers, while other options are used by a smaller portion of respondents.
```
